1. What is unsupervised learning in the context of machine learning?

### Unsupervised Learning in Machine Learning

**Unsupervised learning** is a branch of machine learning that analyzes and clusters unlabeled datasets. These algorithms discover hidden patterns or data groupings without the need for human intervention.

#### Key Characteristics:
* **No Labeled Data:** The input data does not have predefined labels or categories.
* **Pattern Discovery:** It focuses on finding the underlying structure or distribution in the data.
* **Exploratory:** Often used for data exploration to understand the nature of the information.

#### Common Techniques:
1. **Clustering:** Grouping data points into clusters based on similarities (e.g., K-Means, Hierarchical Clustering).
2. **Association:** Discovering rules that describe large portions of data (e.g., 'People who buy X also buy Y').
3. **Dimensionality Reduction:** Reducing the number of variables under consideration (e.g., PCA) while preserving important information.

2.  How does K-Means clustering algorithm work?

### How K-Means Clustering Works

K-Means is an iterative algorithm that partitions a dataset into **K** pre-defined non-overlapping subgroups (clusters). Each data point belongs to the cluster with the nearest mean (centroid).

#### The Step-by-Step Process:

1.  **Initialization:** Choose the number of clusters $K$ and randomly select $K$ data points as the initial centroids.
2.  **Assignment:** Assign each data point to the nearest centroid based on a distance metric (usually Euclidean distance).
3.  **Update:** Calculate the new mean (center) of all data points assigned to each cluster. These new means become the updated centroids.
4.  **Repeat:** Continue repeating the Assignment and Update steps until:
    *   The centroids no longer change significantly.
    *   The assignment of data points to clusters remains the same.
    *   A maximum number of iterations is reached.

#### Goal:
The algorithm aims to minimize the **Within-Cluster Sum of Squares (WCSS)**, also known as inertia, which measures how tightly grouped the clusters are.

3. Explain the concept of a dendrogram in hierarchical clustering.

### Understanding Dendrograms in Hierarchical Clustering

A **dendrogram** is a tree-like diagram that records the sequences of merges or splits during hierarchical clustering. It is the primary tool for visualizing the relationships between data points in this type of unsupervised learning.

#### Key Components:
*   **Leaves:** Each leaf at the bottom represents an individual data point.
*   **Clades (Branches):** The vertical lines represent clusters. The height of the horizontal line connecting two branches indicates the distance (dissimilarity) between those clusters.
*   **Height/Distance:** The y-axis represents the distance metric (e.g., Euclidean distance). A higher horizontal line means the clusters being merged are more different from each other.

#### How to Determine the Number of Clusters:
To choose the number of clusters, you can imagine drawing a horizontal line across the dendrogram at a certain height. The number of vertical lines intersected by this horizontal line tells you the number of clusters at that level of similarity.

#### Summary:
*   **Close proximity** in the tree suggests data points are similar.
*   **Large vertical distances** between merges suggest that the groups are distinct and well-separated.

### K-Means vs. Hierarchical Clustering

5. What are the advantages of DBSCAN over K-Means?

### Advantages of DBSCAN over K-Means

Density-Based Spatial Clustering of Applications with Noise (**DBSCAN**) is a powerful alternative to K-Means that addresses several of its limitations.

#### 1. No Need to Specify $K$
Unlike K-Means, which requires you to pre-define the number of clusters, DBSCAN determines the number of clusters automatically based on the density of the data points.

#### 2. Arbitrary Cluster Shapes
K-Means assumes that clusters are spherical and roughly the same size. DBSCAN can find clusters of any shape (e.g., elongated, circular, or irregular) because it groups points based on local density.

#### 3. Robustness to Outliers
DBSCAN has a built-in notion of **noise**. Any point that is not in a high-density region is labeled as an outlier rather than being forced into a cluster, preventing noise from skewing the cluster centers.

#### 4. Handling Varied Cluster Sizes
DBSCAN can effectively identify clusters of different sizes, as long as the density within the clusters is sufficient according to the specified parameters ($eps$ and $min\_samples$).

#### Summary Comparison:
* **K-Means:** Fast, assumes spherical shapes, requires $K$, sensitive to outliers.
* **DBSCAN:** Handles noise, finds complex shapes, doesn't require $K$, but can struggle with clusters of significantly varying densities.

6. When would you use Silhouette Score in clustering?

### When to Use Silhouette Score in Clustering

The **Silhouette Score** is a metric used to calculate the goodness of a clustering technique. It measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation).

#### Key Uses:
1. **Determining the Optimal Number of Clusters ($K$):** By calculating the silhouette score for different values of $K$, you can identify the value that maximizes the score, indicating the most appropriate clustering structure.
2. **Evaluating Cluster Quality:** It provides a quantitative way to assess how well-separated the resulting clusters are.
3. **Comparing Algorithms:** You can use it to compare the performance of different clustering algorithms (e.g., K-Means vs. Hierarchical) on the same dataset.

#### Interpreting the Score:
The score ranges from **-1 to +1**:
*   **+1:** Indicates that the sample is far away from the neighboring clusters (excellent clustering).
*   **0:** Indicates that the sample is on or very close to the decision boundary between two neighboring clusters (overlapping clusters).
*   **-1:** Indicates that those samples might have been assigned to the wrong cluster.

#### Summary:
While the **Elbow Method** (WCSS) focuses on internal cluster compactness, the **Silhouette Score** provides a more balanced view by considering both compactness and separation.

7. What are the limitations of Hierarchical Clustering?

### Limitations of Hierarchical Clustering

While hierarchical clustering provides a detailed view of data relationships via dendrograms, it has several key limitations:

#### 1. Computational Complexity
Hierarchical clustering is computationally expensive. The time complexity is typically $O(n^2)$ or $O(n^3)$, and the space complexity is $O(n^2)$. This makes it unsuitable for large datasets where $n$ (the number of data points) is in the thousands or more.

#### 2. No Reversibility
Once a merge or split is performed at a specific step, it cannot be undone. This means that if an early merge was suboptimal due to noise or outliers, the error will propagate throughout the rest of the hierarchy.

#### 3. Sensitivity to Noise and Outliers
Like K-Means, hierarchical clustering can be sensitive to outliers. Outliers may either form their own tiny clusters or force other distinct clusters to merge prematurely.

#### 4. Difficulty in Choosing the 'Right' Number of Clusters
Although you don't need to specify $K$ upfront, deciding where to 'cut' the dendrogram can be subjective. There isn't always a clear gap that indicates the most natural number of clusters.

#### 5. Difficulty Handling Different Cluster Shapes and Sizes
Depending on the linkage method used (e.g., Single, Complete, or Ward), the algorithm may struggle to identify clusters with significantly different shapes or densities.

8. Why is feature scaling important in clustering algorithms like K-Means?

### Importance of Feature Scaling in Clustering

Many clustering algorithms, particularly **K-Means** and **Hierarchical Clustering**, rely on calculating the distance (usually Euclidean distance) between data points to determine similarity.

#### 1. Preventing Feature Dominance
If one feature has a range of values from 0 to 1,000,000 (e.g., House Price) and another feature ranges from 0 to 10 (e.g., Number of Rooms), the distance calculation will be almost entirely driven by the feature with the larger range. This effectively ignores the information provided by the smaller-scale feature.

#### 2. Ensuring Equal Contribution
Feature scaling (such as **Standardization** or **Normalization**) ensures that all features contribute equally to the distance metric. This allows the algorithm to find more accurate and meaningful clusters based on the actual relationship between variables rather than their arbitrary units of measurement.

#### 3. Faster Convergence
For iterative algorithms like K-Means, scaling can help the algorithm converge faster. Centroids will move more smoothly toward the local optima when the data is centered and scaled.

#### Common Methods:
*   **Standardization (Z-score scaling):** Transforms data to have a mean of 0 and a standard deviation of 1.
*   **Normalization (Min-Max scaling):** Rescales the data to a fixed range, usually 0 to 1.

9. How does DBSCAN identify noise points?

### How DBSCAN Identifies Noise Points

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) classifies data points into three categories based on two parameters: **eps** (the maximum distance between two points to be considered neighbors) and **min_samples** (the minimum number of points required to form a dense region).

#### The Three Types of Points:

1.  **Core Points:** A point is a core point if it has at least `min_samples` points (including itself) within its `eps` radius.
2.  **Border Points:** A point is a border point if it has fewer than `min_samples` within its `eps` radius, but it is reachable from a core point (i.e., it is within the `eps` radius of a core point).
3.  **Noise Points (Outliers):** A point is a noise point if it is neither a core point nor a border point.

#### Why this is Effective:

*   **Isolation:** Noise points are effectively isolated from high-density regions. Unlike K-Means, which forces every point into a cluster, DBSCAN explicitly leaves these points unassigned.
*   **Parameter Sensitivity:** The identification of noise is directly controlled by `eps` and `min_samples`. Increasing `min_samples` or decreasing `eps` will generally result in more points being classified as noise.

10. Define inertia in the context of K-Means.

### Defining Inertia in K-Means

**Inertia**, also known as the **Within-Cluster Sum of Squares (WCSS)**, is the primary objective function that the K-Means algorithm tries to minimize.

#### Definition:
Inertia measures the sum of squared Euclidean distances between each data point and the centroid of its assigned cluster.

Mathematically, for $K$ clusters, it is represented as:
$$\text{Inertia} = \sum_{i=1}^{K} \sum_{x \in C_i} ||x - \mu_i||^2$$
Where:
*   $C_i$ is the $i^{th}$ cluster.
*   $\mu_i$ is the centroid of cluster $C_i$.
*   $x$ represents the data points within that cluster.

#### Key Characteristics:
1.  **Measure of Compactness:** A lower inertia value indicates that the data points are very close to their respective centroids, meaning the clusters are dense and compact.
2.  **Scale Dependency:** Because it relies on Euclidean distance, inertia is sensitive to the scale of the features (which is why feature scaling is so important).
3.  **The Elbow Method:** Inertia is used in the 'Elbow Method' to find the optimal number of clusters. As $K$ increases, inertia naturally decreases. The 'elbow' is the point where the rate of decrease slows down significantly, suggesting a good balance between the number of clusters and their compactness.

#### Limitation:
Inertia assumes that clusters are convex and isotropic (spherical), so it may not be a reliable metric for clusters with irregular shapes or varying densities.

11. What is the elbow method in K-Means clustering?

### The Elbow Method in K-Means Clustering

The **Elbow Method** is a popular technique used to determine the optimal number of clusters ($K$) for the K-Means algorithm.

#### How it Works:
1.  **Iterate:** Run the K-Means algorithm for a range of values for $K$ (e.g., from 1 to 10).
2.  **Calculate Inertia:** For each value of $K$, calculate the **Inertia** (Within-Cluster Sum of Squares).
3.  **Plot:** Create a line plot with the number of clusters ($K$) on the x-axis and the Inertia on the y-axis.
4.  **Identify the 'Elbow':** As $K$ increases, the inertia will always decrease because the clusters become smaller and more compact. However, there is usually a point where the rate of decrease drops sharply. This point, which looks like an 'elbow' on the graph, suggests the optimal value for $K$.

#### Why Choose the Elbow?
*   **Before the Elbow:** Adding more clusters significantly improves the compactness of the groups (large reduction in inertia).
*   **After the Elbow:** Adding more clusters results in diminishing returns; the improvement in compactness is marginal, and you risk over-partitioning the data.

#### Limitations:
In some datasets, the 'elbow' may not be clearly visible (the curve is smooth), making it subjective. In such cases, other metrics like the **Silhouette Score** are often used in conjunction with the elbow method.

12. Describe the concept of "density" in DBSCAN.

### The Concept of "Density" in DBSCAN

Unlike K-Means which uses distance to centroids, **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) defines clusters as continuous regions of high density.

#### The Two Pillars of Density:

1.  **Epsilon ($\epsilon$ or `eps`):** This is the distance threshold. It defines the neighborhood around a data point. If the distance between two points is less than or equal to `eps`, they are considered neighbors.
2.  **Minimum Samples (`min_samples`):** This is the minimum number of data points required to form a dense region (a cluster).

#### Defining a Dense Region:
A point is in a **dense region** if there are at least `min_samples` within its `eps` radius.

*   **High Density:** If a point meets the `min_samples` criteria, it becomes a **Core Point**, serving as the foundation of a cluster.
*   **Density Reachability:** A point is 'reachable' if it lies within the `eps` distance of a Core Point. This allows clusters to grow into arbitrary shapes as long as Core Points are connected by their dense neighborhoods.
*   **Low Density:** Areas where points are scattered such that they do not meet the `min_samples` requirement and are not reachable from any Core Point are considered **Noise**.

#### Why it Matters:
This density-based approach allows DBSCAN to identify clusters of varying shapes (like a crescent moon or a ring) that distance-to-centroid algorithms would fail to capture.

 14. What does a negative Silhouette Score indicate ?

### What a Negative Silhouette Score Indicates

A **negative Silhouette Score** (between 0 and -1) is a critical indicator of poor clustering quality for specific data points.

#### The Meaning:
*   **Misclassification:** It means that the average distance from a data point to all other points in its own cluster is **greater** than the average distance to points in the nearest neighboring cluster.
*   **Overlap/Incorrect Assignment:** Essentially, the point is more similar to a cluster it *doesn't* belong to than the one it was *assigned* to.

#### Why it Happens:
1.  **Overlapping Clusters:** If clusters are very close or overlapping, points near the boundaries may end up with negative scores.
2.  **Wrong Choice of $K$:** If the number of clusters $K$ is too high or too low, the algorithm might force points into unnatural groupings.
3.  **Outliers:** Outliers that are forced into a cluster by algorithms like K-Means often result in negative silhouette values.
4.  **Inappropriate Distance Metric:** Using a distance metric that doesn't fit the structure of the data can lead to poor assignments.

#### Summary:
If you see many negative scores, it's a sign that the clustering structure is weak and you should consider re-evaluating your preprocessing steps, the number of clusters, or the algorithm itself.

15. Explain the term "linkage criteria" in hierarchical clustering?

### Linkage Criteria in Hierarchical Clustering

In **Hierarchical Clustering**, the **linkage criterion** determines which distance to use between sets of observations (clusters) to decide which clusters should be merged at each step.

#### Common Linkage Types:

1.  **Single Linkage (Nearest Neighbor):**
    *   **Definition:** The distance between two clusters is defined as the minimum distance between any single point in the first cluster and any single point in the second cluster.
    *   **Effect:** Tends to produce long, trailing clusters (the 'chaining' effect).

2.  **Complete Linkage (Farthest Neighbor):**
    *   **Definition:** The distance between two clusters is the maximum distance between any single point in the first cluster and any single point in the second cluster.
    *   **Effect:** Tends to produce more compact, even-sized clusters.

3.  **Average Linkage (UPGMA):**
    *   **Definition:** The distance is the average of all distances between every point in the first cluster and every point in the second cluster.
    *   **Effect:** A compromise between Single and Complete linkage.

4.  **Ward's Linkage:**
    *   **Definition:** Instead of directly measuring distance, Wardâ€™s method minimizes the total within-cluster variance. It merges clusters that lead to the smallest increase in the sum of squared errors (SSE).
    *   **Effect:** Very effective at creating tight, spherical clusters; it is often the default choice in many libraries like Scikit-Learn.

#### Why it Matters:
The choice of linkage can lead to very different results on the same dataset. For example, Ward's linkage is great for finding clusters of similar size, while Single linkage is better for finding elongated or non-spherical shapes but is highly sensitive to noise.

16.  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

### Why K-Means Struggles with Varying Sizes and Densities

K-Means is based on the assumption that clusters are **spherical (isotropic)** and have **similar variances**. When these assumptions are violated, the algorithm often performs poorly.

#### 1. Varying Cluster Sizes
K-Means attempts to minimize the total within-cluster sum of squares. If one cluster is much larger in spatial extent than another, K-Means might 'steal' points from the larger cluster and assign them to a nearby smaller cluster to balance the variances, even if the points are naturally part of the larger group.

#### 2. Varying Cluster Densities
K-Means does not have a concept of density; it only considers distance to centroids. In datasets where one cluster is very dense and another is sparse, the algorithm may split the sparse cluster or merge parts of it with the dense one because it essentially treats all space between centroids equally, regardless of the number of points in those regions.

#### 3. Non-Spherical Shapes
Because K-Means uses a distance-to-centroid approach, it naturally creates 'Voronoi cells' (linear boundaries). If your data forms natural clusters in elongated, U-shaped, or concentric ring patterns, K-Means will fail to capture these shapes and will instead force them into circular partitions.

#### Summary:
For data with these characteristics, density-based algorithms like **DBSCAN** or distribution-based algorithms like **Gaussian Mixture Models (GMM)** are often more appropriate.

17.  What are the core parameters in DBSCAN, and how do they influence clustering?

### Core Parameters in DBSCAN

The performance and results of **DBSCAN** are heavily dependent on two key parameters: **Epsilon ($eps$)** and **Minimum Samples ($min\_samples$)**.

#### 1. Epsilon (eps)
*   **Definition:** The maximum distance between two samples for one to be considered as in the neighborhood of the other.
*   **Influence:**
    *   **Small $eps$:** If $eps$ is too small, a large part of the data will not be clustered and will be considered noise because the 'neighborhoods' are too restricted to form dense regions.
    *   **Large $eps$:** If $eps$ is too large, clusters will merge together, and eventually, the entire dataset might be grouped into a single cluster.

#### 2. Minimum Samples (min_samples)
*   **Definition:** The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.
*   **Influence:**
    *   **Small $min\_samples$:** (e.g., 1 or 2) will result in many small clusters and very little noise. A $min\_samples$ of 1 is essentially hierarchical clustering with single linkage.
    *   **Large $min\_samples$:** Will require higher density to form a cluster. This is useful for noisy datasets as it forces the algorithm to find more significant, dense patterns, but it may also cause smaller, valid clusters to be labeled as noise.

#### How to Choose Them?
*   **Domain Knowledge:** Often the best way to choose these parameters is based on the specific context of the data.
*   **K-Distance Graph:** A common heuristic for choosing $eps$ is to calculate the distance to the $k^{th}$ nearest neighbor for all points (where $k = min\_samples$) and plot them in descending order. The 'elbow' of this plot suggests a good value for $eps$.
*   **Rule of Thumb:** A common starting point for $min\_samples$ is $2 \times \text{dimensions}$, though this varies depending on the amount of noise expected.

18. How does K-Means++ improve upon standard K-Means initialization?

### K-Means++ Initialization

Standard K-Means originally used random initialization for centroids, which often led to poor clustering results because the algorithm is highly sensitive to the starting positions of the centers.

#### The Problem with Random Initialization:
If two initial centroids are placed very close to each other within the same natural cluster, the algorithm may get stuck in a **local optimum**, resulting in clusters that don't represent the true structure of the data.

#### How K-Means++ Improves This:
The core idea of K-Means++ is to spread out the initial centroids. Instead of purely random selection, it uses a probabilistic approach:

1.  **First Centroid:** Choose the first centroid uniformly at random from the data points.
2.  **Distance Calculation:** For each remaining data point, calculate its distance ($D(x)$) to the nearest already-chosen centroid.
3.  **Probabilistic Selection:** Choose the next centroid from the remaining points with a probability proportional to the square of the distance: $P(x) \propto D(x)^2$.
4.  **Repeat:** Repeat steps 2 and 3 until all $K$ centroids have been selected.

#### Key Benefits:
*   **Better Convergence:** By starting with centroids that are already far apart, the algorithm is much more likely to find the global optimum (or a much better local one).
*   **Fewer Iterations:** Because the starting points are better distributed, the algorithm typically requires fewer iterations of the Assignment and Update steps to reach convergence.
*   **Consistency:** It reduces the variance in results across different runs of the algorithm.

19. What is agglomerative clustering?




**Agglomerative clustering** is the most common type of **hierarchical clustering** used to group objects in clusters based on their similarity. It is also known as AGNES (Agglomerative Nesting).

#### The Bottom-Up Approach:
It works using a **bottom-up** strategy:
1.  **Start:** Each data point is initially considered as a single-element cluster (leaf).
2.  **Merge:** At each step, the two clusters that are the most similar (based on a distance metric and linkage criterion) are merged into a single new cluster.
3.  **Repeat:** This process continues iteratively until all points are merged into a single large cluster containing all data points.

#### Key Characteristics:
*   **Dendrogram Representation:** The entire hierarchy of merges is typically visualized using a dendrogram.
*   **Flexibility:** By 'cutting' the dendrogram at different heights, you can decide on the number of clusters after the algorithm has run.
*   **No Global Objective:** Unlike K-Means, which tries to optimize a global objective (inertia), agglomerative clustering makes local merging decisions at each step.

#### When to Use It:
*   When you want to understand the hierarchical relationships between data points.
*   When the number of clusters is not known in advance.
*   When the dataset is small to medium-sized (due to its $O(n^2)$ complexity).

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

### Silhouette Score vs. Inertia

While **Inertia** is the standard objective function for K-Means, the **Silhouette Score** is often considered a more comprehensive metric for model evaluation for several reasons:

#### 1. Internal vs. External Validation
*   **Inertia** only measures **cohesion** (how close points are to their own centroid). It does not care how close clusters are to each other.
*   **Silhouette Score** measures both **cohesion** and **separation**. It evaluates if a point is closer to its own cluster than to the next nearest cluster.

#### 2. Response to the Number of Clusters ($K$)
*   **Inertia** always decreases as you add more clusters, even if those clusters are not meaningful. This forces you to use the 'Elbow Method' to guess the optimal $K$.
*   **Silhouette Score** does not automatically improve with more clusters. It peaks when the clusters are most distinct, providing a much clearer peak for identifying the optimal $K$.

#### 3. Normalized Scale
*   **Inertia** is dependent on the scale of the data and the units of measurement. Its value can range from zero to millions, making it hard to interpret in isolation.
*   **Silhouette Score** is always bounded between **-1 and +1**, providing an absolute measure of quality that can be compared across different datasets and algorithms.

#### 4. Identification of Misclassified Points
*   Inertia provides a single global value for the whole model.
*   The Silhouette Score can be calculated for **individual points**, allowing you to identify specific outliers or points that have been assigned to the wrong cluster (indicated by a negative score).

21.  Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# 2. Apply K-Means clustering
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# 3. Visualize the results
plt.figure(figsize=(10, 7))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=50, c='cyan', label='Cluster 4')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='yellow', marker='*', edgecolors='black', label='Centroids')

plt.title('K-Means Clustering on Synthetic Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

# 1. Generate synthetic data (moons)
X_moons, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# 2. Apply DBSCAN
# eps determines the neighborhood distance, min_samples determines density threshold
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X_moons)

# 3. Visualize the results
plt.figure(figsize=(10, 7))

# Identify unique labels (excluding -1 for noise)
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise points
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X_moons[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], s=50, c=[col],
                label=f'Cluster {k}' if k != -1 else 'Outliers (Noise)')

plt.title('DBSCAN Clustering on Moon-Shaped Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each

47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# 1. Generate noisy blobs
# Increasing cluster_std and adding random points makes it 'noisy'
X_noisy, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# 2. Apply DBSCAN
# We choose eps and min_samples to allow for some points to be labeled as noise
dbscan = DBSCAN(eps=1.0, min_samples=5)
labels = dbscan.fit_predict(X_noisy)

# 3. Identify and count clusters and noise points
unique_labels = np.unique(labels)
n_clusters = len([l for l in unique_labels if l != -1])
n_noise = list(labels).count(-1)

# 4. Print the counts
print(f"Total data points: {len(X_noisy)}")
print(f"Number of clusters identified: {n_clusters}")
print(f"Number of noise points (outliers): {n_noise}")

# Detailed cluster breakdown
for label in unique_labels:
    count = list(labels).count(label)
    if label == -1:
        print(f"Label -1 (Noise): {count} points")
    else:
        print(f"Label {label} (Cluster): {count} points")
```

## Load and Reduce Digits Data with t-SNE

### Subtask:
Load the scikit-learn Digits dataset and apply t-SNE to reduce its dimensionality to 2 components for visualization.


**Reasoning**:
I will import the required libraries, load the Digits dataset, and apply t-SNE to reduce the 64-pixel feature space into 2 components as instructed in the subtask.



```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE

# 1. Load the digits dataset and extract feature matrix
digits = load_digits()
X_digits_agg = digits.data

# 2. Instantiate TSNE with n_components=2 and recommended settings
tsne_agg = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto')

# 3. Apply fit_transform to obtain the 2D representation
X_tsne_agg = tsne_agg.fit_transform(X_digits_agg)

# 4. Print the shape to verify the transformation
print(f"Original shape: {X_digits_agg.shape}")
print(f"Reduced t-SNE shape: {X_tsne_agg.shape}")
```

## Apply Agglomerative Clustering

### Subtask:
Perform Agglomerative Clustering on the 2D t-SNE reduced data using 10 clusters to represent the 10 digit classes (0-9).


**Reasoning**:
I will perform Agglomerative Clustering on the 2D t-SNE reduced digits data using 10 clusters and store the resulting labels as requested.



```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

# 1. Instantiate AgglomerativeClustering with 10 clusters
agg_model = AgglomerativeClustering(n_clusters=10)

# 2. Fit and predict cluster labels on the 2D t-SNE data
agg_tsne_labels = agg_model.fit_predict(X_tsne_agg)

# 3. Print the first 20 labels and unique labels to verify
print(f"First 20 Agglomerative cluster labels:\n{agg_tsne_labels[:20]}")
print(f"Unique labels found: {np.unique(agg_tsne_labels)}")
```

**Reasoning**:

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create the scatter plot using the 2D t-SNE data and Agglomerative labels
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne_agg[:, 0], X_tsne_agg[:, 1], c=agg_tsne_labels, cmap='tab10', alpha=0.6, s=30)

# 2. Add axis labels, title, and legend
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('Agglomerative Clustering of Digits (t-SNE Reduced)')

# 3. Create a legend to distinguish between the 10 clusters
legend1 = plt.legend(*scatter.legend_elements(), title='Clusters', loc='best', bbox_to_anchor=(1, 1))
plt.gca().add_artist(legend1)

# 4. Display the plot
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
